```python
import pandas as pd
from datetime import datetime
import re
import os
```

```python
# tabela historica de infestação

#subir para a pasta raiz do projeto
up_dir = "../"

#acessar caminho do diretorio raw
raw_dir ="sucoslake/trusted/"

#nome do arquivo que contem dos dados brutos
file = "trusted_infestacao.parquet"

data_path = up_dir+raw_dir+file

historico = pd.read_parquet(data_path)
```

```python
# dados coletados no powerapps

#subir para a pasta raiz do projeto
up_dir = "../"

#acessar caminho do diretorio raw
raw_dir ="sucoslake/trusted/"

#nome do arquivo que contem dos dados brutos
file = "trusted_infestacao_powerapps.parquet"

data_path = up_dir+raw_dir+file

powerapps = pd.read_parquet(data_path)
```

```python
# Mantendo os registros mais recentes de cada ID - premissa adotada
historico = historico.sort_values('dh_insercao_trs').groupby('ID', as_index=False).last()
powerapps = powerapps.sort_values('dh_insercao_trs').groupby('ID', as_index=False).last()
```

```python
powerapps.drop('ID', axis=1, inplace=True)
historico.drop('ID', axis=1, inplace=True)
```

```python
df = pd.concat([historico, powerapps], ignore_index=True)
```

```python
# Acesso ao diretorio dos dados trusted

#subir para a pasta raiz do projeto
up_dir = "../"

#acessar caminho do diretorio trusted
raw_dir ="sucoslake/refined/"

#nome do arquivo que contem dos dados brutos
file = "refined_infestacao.parquet"

#caminho completo
refined_path = up_dir+raw_dir+file

if os.path.exists(refined_path) is False: 
    df.to_parquet(refined_path, engine="pyarrow", compression="snappy")
    print("Charge Full")
    
else:
    df_existente = pd.read_parquet(refined_path)
    df_existente = df_existente.sort_values('dh_insercao_trs').groupby(['dt_colheita', 'nm_fazenda', 'cd_talhao', 'ds_fruta'], as_index=False).last()
        
    # Criar um flag de alteração comparando todos os campos exceto a chave
    key = ['dt_colheita', 'nm_fazenda', 'cd_talhao', 'ds_fruta']
    compare_cols = [
    'cd_loc',
    'ds_sinaisPraga',
    'pc_infestacao',
    'nm_resp_campo'
    ]

    # Merge para detectar alterações
    df_merged = df.merge(df_existente, on=key, how='left', suffixes=('_novo','_old'))
    
    # Verifica se algum campo mudou
    mask_alterado = (df_merged[[f"{c}_novo" for c in compare_cols]] != df_merged[[f"{c}_old" for c in compare_cols]].values).any(axis=1)
    
    # Seleciona apenas os registros alterados
    df_para_inserir = df[mask_alterado]
        
    # Adiciona os registros alterados ao histórico
    df_final = pd.concat([df_existente, df_para_inserir], ignore_index=True)
    
    # Salvar arquivo parquet
    #df_final.to_parquet(refined_path, engine="pyarrow", compression="snappy")
    print(f"Insert: {refined_path}")
    df_final
```
